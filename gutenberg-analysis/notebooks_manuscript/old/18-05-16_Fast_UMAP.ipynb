```python
import numpy as np
import pandas as pd
import umap

from scipy.sparse import dok_matrix, csc_matrix

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_features = 15000
n_samples = 2000
alpha = 0.001 ### feature density

data = np.random.randn(n_samples, n_features)
data_sparse = dok_matrix((n_samples, n_features))
for sample in range(n_samples):
    for feature in np.random.choice(range(n_features), int(alpha*n_features), replace=False):
        data_sparse[sample, feature] = np.random.normal()
        
metadata = np.random.choice([0, 1, 2], size=n_samples)
```

```python
tdata = umap.UMAP().fit_transform(data_sparse)
```
